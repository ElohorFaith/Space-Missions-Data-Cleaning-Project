```python
data['Time'].head()
```

```python
# deal with Time missing values
# Drop rows where Time is missing
data = data.dropna(subset=['Time'])

# Reset the index if you want a clean sequential index
data = data.reset_index(drop=True)
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values")
missing_values
```

```python
# Check for Duplicates
duplicates = data.duplicated()
data[duplicates]
```

```python
# Remove Duplicates
data.drop_duplicates(inplace = True)
```

```python
# Check for Duplicates
duplicates = data.duplicated()
data[duplicates]
```

```python
# standardising the Company column
data['Company'].value_counts().head(20)
```

```python
# remove spaces and turn the data in the company column to title case
data['Company'] = data['Company'].str.strip()
data['Company'] = data['Company'].str.title()
```

```python
# standardising the Date column
data['Date'].head(20)
```

```python
# ensures the date column is turned to string datatype before removing extra spaces
data['Date'] = data['Date'].astype(str).str.strip()
```
