```python
import pandas as pd
import numpy as np
```

```python
# load data
data = pd.read_excel("C:/Users/ElohorEfere/Documents/DATA ANALYTICS/PYTHON/space_missions.xlsx")
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values")
missing_values
```

```python
#Fill the Missing Price rows with 0 
data['Price'] = data['Price'].fillna(0)
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values")
missing_values
```

```python
data.dtypes
```

```python
# convert Time column to DateTime datatype (errors='coerce' means any weird or missing value becomes NaT instead of crashing.)
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M:%S %p', errors='coerce')
```

```python
# Check to see the new datatype after the conversion of the Time column
data.dtypes
```

```python
# fill the missing time values with 00:00:00
data['Time'] = data['Time'].fillna(pd.to_datetime('00:00:00'))
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values")
missing_values
```

```python
# Drop the Time_converted column that was created as a result of the clean
data = data.drop(columns=['Time_converted'])
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values")
missing_values
```

```python
# Check for Duplicates
duplicates = data.duplicated()
data[duplicates]
```

```python
# Remove Duplicates
data.drop_duplicates(inplace = True)
```

```python
# Check for Duplicates
duplicates = data.duplicated()
data[duplicates]
```
